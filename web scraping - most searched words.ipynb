```python
# importing necessary library for scraping.
```

```python
# pandas for dataframe 
import pandas as pd

# requests for connecting url to machine
import requests

# bs4 that is BeautifulSoup,a Python library for pulling data out of HTML and XML files.
import bs4
```

```python
# url of particular website 
```

```python
url = 'https://www.infidigit.com/blog/top-google-searches/'
```

```python
# connecting to url 
```

```python
res = requests.get(url)
```

```python
# beautifulsoup along with '.text' to get text from html and 'html.parser' for parsing text.
```

```python
soup = bs4.BeautifulSoup(res.text,'html.parser')
```

```python
soup
```

```python
# creating a variable 'table' for finding the particular class tag , id tag and thead tag.
```

```python
table = soup.find(class_="tablepress tablepress-id-14",id='tablepress-14-no-2').find('thead')
```

```python
table
```

```python
# empty header for storing values
```

```python
header = []
```

```python
#iterating through each 'th' tags
#.text.strip() to get text and stripping any non visible line like spaces line break
```

```python
for i in table.find_all('th'):
    title = i.text.strip()
    header.append(title)
```

```python
header
```

```python
# setting dataframe and making header as a column
```

```python
df = pd.DataFrame(columns=header)
```

```python
df
```

```python
# creating another variable 'table2' for finding the particular class tag , id tag and tbody tag as all the data lies within it.
```

```python
table2 = soup.find(class_="tablepress tablepress-id-14",id='tablepress-14-no-2').find('tbody')
```

```python
table2
```

```python
# iterating through each 'tr' tags to select tr tags 
# iterating through each 'td' tags to select td tags and grabbing its text.
# length = len(df) for selecting length of dataframe
# df.loc[length] = row_data for merging values to dataframe
```

```python
for row in table2.find_all('tr')[:]:
    data = row.find_all('td')
    row_data = [td.text for td in data]
    print(row_data)
    length = len(df)
    #print(length)
    
    df.loc[length] = row_data
    #print(df.loc[length])
```
